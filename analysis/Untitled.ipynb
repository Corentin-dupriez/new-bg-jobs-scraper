```python
import polars as pl
import os
import duckdb
```

```python
parent_folder = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
```

```python
database = os.path.join(parent_folder,'jobs_scraping', 'jobs.duckdb')
```

```python
con = duckdb.connect(database)
```

```python
con.execute("SHOW TABLES").df()
```

```python
pl_df = pl.from_arrow(
    con.execute("SELECT * FROM fact_listings").arrow()
)
```
