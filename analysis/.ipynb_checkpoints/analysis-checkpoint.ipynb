```python
import polars as pl
import os
import duckdb
```

```python
parent_folder = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
```

```python
database = os.path.join(parent_folder,'jobs_scraping', 'jobs.duckdb')
```

```python
con = duckdb.connect(database)
```

```python
con.execute("SHOW TABLES").df()
```

```python
fact_listings = pl.from_arrow(
    con.execute("SELECT * FROM fact_listings").arrow()
)
```

```python
fact_listings.filter(pl.col('min_salary') > "5000").head(5)
```

```python
dim_company = pl.from_arrow(
    con.execute('SELECT * FROM dim_company').arrow()
)
```

```python
dim_company.head(4)
```

```python
fact_listings.join(dim_company, on='company_id', how='left').select(['job_title','company_name'])
```

```python
dim_categories = pl.from_arrow(
    con.execute('SELECT d.category_id, d.category_name, b.listing_id FROM dim_categories d join bridge_listing_category b on b.category_id = d.category_id').arrow()
)
```
