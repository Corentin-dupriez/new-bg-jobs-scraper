```python
import polars as pl
import os
import duckdb
```

```python
parent_folder = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
```

```python
database = os.path.join(parent_folder,'jobs_scraping', 'jobs.duckdb')
```

```python
con = duckdb.connect(database)
```

```python
con.execute("SHOW TABLES").df()
```

```python
fact_listings = pl.from_arrow(
    con.execute("SELECT * FROM fact_listings").arrow()
)
```

```python
listings_with_salary = fact_listings.filter(pl.col('min_salary') != 'Not communicated',
                                           pl.col('min_salary') != '',
                                           pl.col('max_salary') != 'Not communicated').select(
    pl.col('company_id'),
    pl.col('job_title'),
    pl.col('location_id'),
    pl.col('min_salary').cast(pl.Int32),
    pl.col('max_salary').cast(pl.Int32),
    pl.col('listing_url'),
    pl.col('description'),
    pl.col('date_posted_id'),
    pl.col('date_scraped_id')
)
```

```python
listings_with_salary.group_by(
    pl.col('job_title')
).agg(
    pl.col('min_salary').mean().alias('avg_min_salary'),
    pl.col('max_salary').mean().alias('avg_max_salary')
).head(10)
```
